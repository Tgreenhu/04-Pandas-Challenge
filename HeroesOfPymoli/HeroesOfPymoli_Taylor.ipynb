```python
# Dependencies
import pandas as pd
```

```python
# Import CSV & save to a variable
file = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file)
```

```python
# Look at a snippet of our data
purchase_data.head()
```

```python
# Find total num of players then set it as a DF
unique_total_players = purchase_data["SN"].unique()
total_players = len(unique_total_players)

# Player Count DataFrame
number_of_players = pd.DataFrame({"Total Players": [total_players]})

number_of_players
```

```python
# Find the num of unique items
unique_items = len(purchase_data["Item Name"].unique())
# Find the avg of purchased items
average_price = purchase_data["Price"].mean()
# Find the total num of purchases
total_purchases = purchase_data["Purchase ID"].count()
# Find total revenue made
total_revenue = purchase_data["Price"].sum()

# Purchasing Analysis (Total) DataFrame
total_purchasing_analysis = pd.DataFrame({"Number of Unique Items": [unique_items], "Average Price": [average_price],
                                          "Number of Purchases": [total_purchases], "Total Revenue": [total_revenue]})

# Style Formatting
total_purchasing_analysis["Average Price"] = total_purchasing_analysis["Average Price"].map("${:,.2f}".format)
total_purchasing_analysis["Total Revenue"] = total_purchasing_analysis["Total Revenue"].map("${:,}".format)

total_purchasing_analysis
```

```python
# Sort our df by gender and find our totals
genders = purchase_data.groupby(['Gender'])
total_genders = genders["SN"].nunique()
percentage = (total_genders/total_players)

# Gender Demographics Data Frame
gd = pd.DataFrame({'Total Count': total_genders, 'Percentage of Players': percentage})

# Style Formatting
gender_demographics = gd.sort_values(['Total Count'], ascending=False)
gender_demographics = gender_demographics.rename_axis([""])
gender_demographics['Percentage of Players'] = gender_demographics['Percentage of Players'].map("{:.2%}".format)
gender_demographics
```

```python
# Future sort our df by gender and find our totals
purchase_count = genders['Purchase ID'].count()
average_purchase = genders['Price'].mean()
total_value = genders['Price'].sum()
avg_per_person_per_gender = total_value/total_genders

# Purchasing Analysis (Gender) Data Frame
purchasing_analysis = pd.DataFrame({'Purchase Count': purchase_count, 'Average Purchase Price': average_purchase,
                                    'Total Purchase Value': total_value, 'Avg Total Purchase Per Person': avg_per_person_per_gender})

# Style Formatting
purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis["Avg Total Purchase Per Person"] = purchasing_analysis["Avg Total Purchase Per Person"].map("${:,.2f}".format)

purchasing_analysis
```

```python
#Finding min (7) & max (45) ages
max_age = purchase_data["Age"].max()
min_age = purchase_data["Age"].min()

#Creating our bins/labels & cut Age Range column
bins = [5, 9, 14, 19, 24, 29, 34, 39, 50]
groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data['Age Range'] = pd.cut(purchase_data["Age"], bins, labels=groups, include_lowest=True)

# Sort our df by age and find our totals
ad = purchase_data.groupby('Age Range')
age_range = ad["SN"].nunique()
age_percentage = (age_range/total_players).astype(float)

# Age Demographics DataFrame
age_demographics = pd.DataFrame({'Total Count': age_range, 'Percentage of Players': age_percentage})
age_demographics = age_demographics.rename_axis([""])
age_demographics['Percentage of Players'] = age_demographics['Percentage of Players'].map("{:.2%}".format)

age_demographics
```

```python
# Find necessary calculations (purchase count, avg. purchase price, avg. purchase total per person)
purchase_count_2 = ad["Purchase ID"].count()
avg_purchase_price = ad["Price"].mean()
total_purchase_price = ad["Price"].sum()
avg_total_person = total_purchase_price/age_range

# Purchasing Analysis (Age) DF
purchasing_analysis_df = pd.DataFrame({'Purchase Count': purchase_count_2,
                                       'Average Purchase Price': avg_purchase_price,
                                       'Total Purchase Value': total_purchase_price,
                                       'Avg Total Purchase Per Person': avg_total_person})

# Style Formatting
purchasing_analysis_df["Average Purchase Price"] = purchasing_analysis_df["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_df["Total Purchase Value"] = purchasing_analysis_df["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_df["Avg Total Purchase Per Person"] = purchasing_analysis_df["Avg Total Purchase Per Person"].map("${:,.2f}".format)


purchasing_analysis_df
```

```python
# Group data by "SN" then find neccessary calcs (purchases, avg purchase price, total purchase value)
top_spenders = purchase_data.groupby("SN")
ts_purchase_count = top_spenders['Purchase ID'].count()
ts_avg_purchase_price = top_spenders['Price'].mean()
ts_total_purchase_value = top_spenders['Price'].sum()

# Top Spenders DataFrame
top_spenders_df = pd.DataFrame({'Purchase Count': ts_purchase_count,
                                'Average Purchase Price': ts_avg_purchase_price,
                                'Total Purchase Value': ts_total_purchase_value})

# Sort our df by decending total purchase value
sorted_ts_df = top_spenders_df.sort_values("Total Purchase Value", ascending=False)

# Style Formatting
sorted_ts_df['Average Purchase Price'] = sorted_ts_df['Average Purchase Price'].map("${:,.2f}".format)
sorted_ts_df['Total Purchase Value'] = sorted_ts_df['Total Purchase Value'].map("${:,.2f}".format)

sorted_ts_df.head()
```

```python
# Create df & group it by specific items...
# ...then find neccessary calcs (purchase count, average item price, and total purchase value)
most_popular = purchase_data[["Item ID", "Item Name", "Price"]]
most_popular_grouped = most_popular.groupby(["Item ID", "Item Name"])
mp_purchase_count = most_popular_grouped["Price"].count()
mp_item_price = most_popular_grouped["Price"].mean()
mp_purchase_value = most_popular_grouped["Price"].sum()

# Most Popular Items DataFrame
most_popular_items = pd.DataFrame({'Purchase Count': mp_purchase_count,
                                   'Item Price': mp_item_price,
                                   'Total Purchase Value': mp_purchase_value})

# Sort our df by decending purchase count
sorted_mpi_df = most_popular_items.sort_values("Purchase Count", ascending=False)

# Style Formatting
sorted_mpi_df['Item Price'] = sorted_mpi_df['Item Price'].map("${:,.2f}".format)
sorted_mpi_df['Total Purchase Value'] = sorted_mpi_df['Total Purchase Value'].map("${:,.2f}".format)

sorted_mpi_df.head()







```

```python
# Sort our df above by decending total purchase value
most_profitable = most_popular_items.sort_values("Total Purchase Value", ascending=False)

# Style Formatting
most_profitable['Item Price'] = most_profitable['Item Price'].map("${:,.2f}".format)
most_profitable['Total Purchase Value'] = most_profitable['Total Purchase Value'].map("${:,.2f}".format)

most_profitable.head()
```
